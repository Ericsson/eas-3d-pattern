### Plot interactive antenna pattern

```python
from eas_3d_pattern import SAMPLE_JSON, AntennaPattern

antenna_pattern = AntennaPattern(SAMPLE_JSON[0])  # load in a sample JSON file
```

We simply use the *plot* method on the *antenna_pattern*, which will give us an interactive heatmap of the normalized power pattern in dB

```python
antenna_pattern.plot()
```

We can also plot other components.

Note: By default the range is heatmap range is always between 0 and -30.

```python
antenna_pattern.plot(component_name="P_co_dB")
```

Let's customize the plot even more and save it

```python
fig = antenna_pattern.plot(component_name="P_tp_lin", show_fig=False)
fig.data[0].update(zmin=0, zmax=1)  # change the z limits
fig.show()  # render it
```

We can also export the figure we just created as .html

```python
fig.write_html("antenna_pattern_file1.html")
```
